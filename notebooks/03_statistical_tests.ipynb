# 03. Statistical Hypothesis Testing

This notebook runs standard non-parametric repeated measures hypothesis testing (Friedman test) and pairwise Wilcoxon signed-rank tests to evaluate the statistical significance of RL model cost improvements over other structures.

```python
import os
import sys
sys.path.append(os.path.abspath(".."))

from src.evaluation.statistics import run_friedman_and_wilcoxon_tests

rl_files = {
    "MAPPO": "All Result - MAPPO.xlsx",
    "SA-T":  "All Result-SA-T.xlsx",
    "SA-NT": "All Result-SA-NT.xlsx",
}
out_file = "All Result-OUT.xlsx"

results = run_friedman_and_wilcoxon_tests(rl_files, out_file)

print("=========================================================================")
print("📊 GLOBAL FRIEDMAN TEST RESULTS (Repeated Measures Repeated across 50 Seeds)")
print("=========================================================================")
print(f"Friedman Chi-Squared Statistic: {results['friedman']['statistic']:.4f}")
print(f"p-value: {results['friedman']['p_value']:.4e}")

print("\n=========================================================================")
print("📊 POST-HOC PLANNED PAIRWISE WILCOXON SIGNED-RANK TESTS (Holm Correction)")
print("=========================================================================")
for res in results["pairwise"]:
    print(f"{res['comparison']} | Stat: {res['wilcoxon_stat']:.2f} | Raw p: {res['raw_p']:.4e} | Holm p: {res['corrected_p']:.4e} | Significant: {res['significant']}")

print("\n=========================================================================")
print("📊 ADDITIONAL ROBUSTNESS MODEL-LEVEL CHECKS (n=5 Training Seeds)")
print("=========================================================================")
for res in results["model_level"]:
    print(f"{res['comparison']} | Paired t-p: {res['t_p']:.4f} | Wilcoxon p: {res['wilcoxon_p']:.4f}")
```
